```python
import numpy as np
import pandas as pd
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
#putting movies data on 'movies' dataframe
movies = pd.read_csv('main_data.csv')
```

```python
movies.head(5)
```

```python
movies.columns
```

```python
movies = movies.loc[:,['Title','Genre','Tags','Languages','Director','Writer','Actors','Summary']]
```

```python
movies.head(10)
```

```python
movies['Title'] = movies['Title'].replace(np.nan, 'unknown')
movies['Genre'] = movies['Genre'].replace(np.nan, 'unknown')
movies['Tags'] = movies['Tags'].replace(np.nan, 'unknown')
movies['Languages'] = movies['Languages'].replace(np.nan, 'unknown')
movies['Director'] = movies['Director'].replace(np.nan, 'unknown')
movies['Writer'] = movies['Writer'].replace(np.nan, 'unknown')
movies['Actors'] = movies['Actors'].replace(np.nan, 'unknown')
movies['Summary'] = movies['Summary'].replace(np.nan, 'unknown')
```

```python
movies
```

```python
movies['Genre'] = movies['Genre'].str.replace(',', ' ')
movies['Tags'] = movies['Tags'].str.replace(',', ' ')
movies['Languages'] = movies['Languages'].str.replace(',', ' ')
movies['Writer'] = movies['Writer'].str.replace(',', ' ')
movies['Actors'] = movies['Actors'].str.replace(',', ' ')
movies['Summary'] = movies['Summary'].str.replace(',', ' ')
```

```python
movies
```

```python
movies['Title'] = movies['Title'].str.lower()
```

```python
movies['comb'] = movies['Title'] + ' ' + movies['Genre'] + ' '+ movies['Tags'] + ' '+ movies['Languages'] +' ' + movies['Director']+' ' +movies['Writer'] + ' '+ movies['Actors'] +' ' + movies['Summary']
```

```python
movies.to_csv('main_data.csv',index=False)
```
